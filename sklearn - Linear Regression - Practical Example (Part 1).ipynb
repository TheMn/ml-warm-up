# Practical example

## Importing the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

## Loading the raw data

```python
raw_data = pd.read_csv('1.04. Real-life example.csv')
raw_data.head()
```

## Preprocessing

### Exploring the descriptive statistics of the variables

```python
raw_data.describe(include='all')
```

### Determining the variables of interest

axis = 0 means rows and axis = 1 stands for columns

```python
data = raw_data.drop(['Model'],axis=1)
data.describe(include='all')
```

### Dealing with missing values

```python
data.isnull().sum()
```

```python
data_no_mv = data.dropna(axis=0)
```

```python
data_no_mv.describe(include='all')
```

### Exploring the PDFs

```python
sns.distplot(data_no_mv['Price'])
```

For optimal results we would be looking for a normal distribution

### Dealing with outliers

Outliers: observations that lie on abnormal distance from other observations in the data

```python
q = data_no_mv['Price'].quantile(0.99)
data_1 = data_no_mv[data_no_mv['Price']<q]
data_1.describe(include='all')
```

While max price value still far away from the mean, it is acceptably closer

```python
sns.distplot(data_1['Price'])
```

```python
sns.distplot(data_no_mv['Mileage'])
```

```python
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage']<q]
```

```python
sns.distplot(data_2['Mileage'])
```

```python
sns.distplot(data_no_mv['EngineV'])
```

An interval where engine volume should normally fall is 0.6 to 6.5

A common way to label missing values is by assigning 99.99, but it is a bad idea to label values in such ways as it is hard for other users of the data to distinguish them from true values.

```python
data_3 = data_2[data_2['EngineV']<6.5]
```

```python
sns.distplot(data_3['EngineV'])
```

```python
sns.distplot(data_no_mv['Year'])
```

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year']>q]
```

```python
sns.distplot(data_4['Year'])
```

Currently the indices refer to all the data (including the observations we removed) but we want them to describe only the data userful for analysis.

```python
data_cleaned = data_4.reset_index(drop=True)
```

```python
data_cleaned.describe(include='all')
```

Overall we've deleted ~250 observations but they were the problematic ones.

## Checking the OLS assumptions

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
ax3.set_title('Price and Mileage')


plt.show()
```

We know we should not run a linear regression in this case. We should first transform one or more variables.

```python
sns.distplot(data_cleaned['Price'])
```

### Relaxing the assumptions

Log transformations are especially useful when facing exponential relationships.

```python
log_price = np.log(data_cleaned['Price'])
data_cleaned['log_price'] = log_price
data_cleaned
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['log_price'])
ax1.set_title('Log Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['log_price'])
ax2.set_title('Log Price and EngineV')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['log_price'])
ax3.set_title('Log Price and Mileage')


plt.show()
```

```python
data_cleaned = data_cleaned.drop(['Price'],axis=1)
```

We will discuss the residuals after the regression is created.

### Multicollinearity

```python
data_cleaned.columns.values
```

One of the best ways to check for multicollinearity is through VIF (variance inflation factor)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
```

```python
vif
```

You'll rarely find data so perfect that all features have VIF < 5

```python
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
```
